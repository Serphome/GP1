## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия формируется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае индивидуальная оценка участника становится равной общей оценке команды за проект (т.е. в идеальном случае — вторая оценка равна первой оценке).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(04032026)`.

```python
fig = px.box(data_without_del, y='release year') #поменял на plotly
fig.show()
```

Можно заметить, некоторая часть шоу выпала за границы. Посмотрим, что это за значения, и выберем наши дальнейшие действия - считаем их выбросами, или нет.

```python
Q1 = data_without_del['release year'].quantile(0.25)
Q3 = data_without_del['release year'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Нижняя граница =", lower_bound)
print("Верхняя граница =", upper_bound)
```

```python
outliers = data_without_del[
    (data_without_del['release year'] < lower_bound) |
    (data_without_del['release year'] > upper_bound)
]

print("Количество выбросов:", len(outliers))
outliers
```

По результатам boxplot видно, что в данных есть выбросы. ##Анализ распредления жанров среди тайтлов

Проанализируем наш датасет и узнаем, как распределены жанры по среднему рейтингу пользователей.

Для это сделаем небольшой FA. Мы решили разбить жанры с помощью OHE, чтобы вынести каждую категорию в свой столбец отдельно. Изначально жанры представлены как список строк, разобьем их и применим кодирование.

```python
import ast

data_new['genres'] = data_new['genres'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else []) #без такой штуки не работает так как там есть 'NaN'
# либо можно сразу шоу без жанров удалить и по идее норм без нее будет работать
# https://stackoverflow.com/questions/75011487/use-ast-literal-eval-on-all-columns-of-a-pandas-dataframe
```

```python
categorical_subset = (data_new['genres'].explode()) #раскрываем списки жанров
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.explode.html
categorical_subset = pd.get_dummies(categorical_subset)
features = pd.concat([data_new, categorical_subset], axis=1)
features
```

Отлично, теперь узнаем среднюю оценку по жанрам

```python
genre_list = categorical_subset.columns
mean_ratings = {}
for genre in genre_list:
    mean_ratings[genre] = features[features[genre] == 1]['user rating score'].mean()
mean_ratings_df = (pd.DataFrame(mean_ratings.items(), columns=['genre', 'mean_rating']).sort_values(by='mean_rating', ascending=False))
mean_ratings_df
```

Для того, чтобы оценивание было более объективным и не страдало из-за недостаточной выборки, мы посмотрим кол-во вхождений каждого жанра в наш датасет.

```python
counts = {}

for genre in genre_list:
    counts[genre] = features[genre].sum()
counts_df = pd.DataFrame(counts.items(), columns=['genre', 'count'])

filtered = mean_ratings_df.merge(counts_df, on='genre')
filtered
```

Видим, что такие жанры как music, sport, history, western, war, встречаются намного реже, относительно других жанров, давайте для более объективной оценки их уберем.

```python
filtered = filtered[filtered['count'] > 10].sort_values(by='mean_rating', ascending=False)
filtered
```

Визуализируем полученные результаты и сделаем вывод

```python
fig = px.scatter(filtered, x='genre', y='mean_rating', title='Средний рейтинг по жанрам', size='count', color='mean_rating')
fig.update_layout(template='plotly_white')
```

В результате анализа получается, что самые самый популярный и высоко оцененный жанр - это драма, фильмов снимают много и рейтинг достачно высокий, скорее всего это связано с тем, что то снять такое кино дешевле, чем крутой триллер или крайм, но из-за того, что тема драматургии интересна и вызывает отклик и эмоции у большого количества людей - оценки такие высокие, несмотря на то что снимают очень много подобного кино.

Так же можно отметить, что в самом низу по оценкам находятся "семейные жанры", они доступны даже большему кругу людей, но оценены ниже той же самой драмы из чего можно сделать вывод, что для людей важнее пережитые эмоции полученные при просмотре кино и комедии, анимационные фильмы и семеное кино больше расценивается как разгрузочный и более проходной контент.

На самом деле это далеко не все выводы, которые можно сделать из данного анализа.

##Анализ различия рейтингов Netflix и IMDb

Проанализируем наш датасет на различие рейтингов imdb и netflix

```python
ratings_df = data_new[['title', 'user rating score', 'imdb_score', 'imdb_votes', 'type', 'release year']]

ratings_df = ratings_df.rename(columns={'user rating score': 'netflix_score'})
ratings_df = ratings_df.dropna(subset=['netflix_score', 'imdb_score'])

ratings_df['netflix_score'] = ratings_df['netflix_score'] / 10

ratings_df
```

```python
ratings_df[['netflix_score', 'imdb_score']].describe()
```

Видно, что среднее значение рейтингов у imdb ниже, примерно на 1 пункт. То есть пользователи на imdb оценивают проекты строже чем на netflix. Это также видно по минимальной границе рейтинга: на imdb есть фильм с оценкой 3.9, на нетфликсе худший фильм имеет рейтинг 5.6. При этом максимальные значения отличаются не так сильно - лучшие фильмы хорошой оценивают на обоих площадках, но на нетфликсе лояльность аудитории все равно выше.

```python
ratings_df['diff'] = abs(ratings_df['imdb_score'] - ratings_df['netflix_score'])
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=ratings_df['imdb_score'],
    y=ratings_df['netflix_score'],
    mode='markers',
    marker=dict(size=10, color=ratings_df['diff'], colorscale='spectral'),
    text=ratings_df['title'],
    name='Оценки фильмов',
    hovertemplate = "<b>%{text}</b><br>IMDB: %{x}<br>Netflix: %{y}" #Это для красивого окошка при наведении на точку https://plotly.com/python/hover-text-and-formatting/
))

fig.add_trace(go.Scatter(
    x=[4, 10],
    y=[4, 10],
    mode='lines',
    name='Идеальная корреляция',
    line=dict(color='black', width=2, dash='dash')
))

fig.update_layout(
    title='Сравнение рейтингов Netflix и IMDB',
    xaxis_title='IMDB рейтинг',
    yaxis_title='Netflix рейтинг',
    showlegend=True
)


fig.show()
print('Корреляция рейтингов:', ratings_df['imdb_score'].corr(ratings_df['netflix_score']).round(2))
```

Хоть и числовая корреляция достаточно слабая - 0.33, по графику явно можно заметить зависимости данных. Точек выше пунктирной линии намного больше, чем точек ниже этой линии, так как аудитория нетфликса оценивает фильмы намного лояльнее. Также заметно что почти все данные выстроены по линиям, парралельным прямой y=x (то есть это графики y=x + c), особенно около этой прямой. Поэтому возможно, есть другие параметры, влияющие на расхождение рейтингов.

Посмотрим на количество голосов и разницу в рейтинге по фильмам. Возможно, чем больше голосов, тем более похожими будут рейтинги.

```python
fig = make_subplots(rows=1, cols=2, subplot_titles=('Обычный график', 'Логарифмированный график'))
ratings_df = ratings_df.sort_values(by='imdb_votes')
fig.add_trace(go.Scatter(
    x=ratings_df['imdb_votes'],
    y=ratings_df['diff'],
    mode='markers',
    name='Разница рейтингов',
    marker=dict(size=10),
    text=ratings_df['title'],
    hovertemplate='<b>%{text}</b><br>Голоса: %{x}<br>Разница: %{y}'), row=1, col=1
)

fig.add_trace(go.Scatter(
    x=np.log(ratings_df['imdb_votes']),
    y=ratings_df['diff'],
    mode='markers',
    name='Разница рейтингов',
    marker=dict(size=10),
    text=ratings_df['title'],
    hovertemplate='<b>%{text}</b><br>Голоса: %{x}<br>Разница: %{y}'), row=1, col=2
)

fig.update_xaxes(title_text="Количество голосов")
fig.update_yaxes(title_text="Разница в рейтинге")


fig.update_layout(
    title='Сравнение разницы рейтингов по количеству голосов',
    showlegend=False
)
fig.show()
```

Хоть и по  первому графику видно, что верхняя граница разницы в рейтинге уменьшается с увеличением количества голосов, 100 процентную зависимость доказать нельзя: слишком мало данных. Большинство тайтлов имееют малое количество голосов, поэтому я создал график, логарифмированный по оси x, и тут зависимость уже пропадает, что говорит о слабой корреляции количества голосов и рейтинга.

Теперь посмотрим на зависимость среднего рейтинга от типа контента

```python
gr_df = ratings_df.groupby('type')[['imdb_score', 'netflix_score']].mean().reset_index()
fig = px.bar(gr_df, x='type', y=['imdb_score', 'netflix_score'], barmode='group', title='Средние рейтинги по типу тайтла')

fig.show()
```

На графике различие в рейтинге между сериалами значительно меньше чем между фильмами. Это интересное наблюдение: аудитория, которая смотрит сериалы, скорее всего использует нетфликс, и ставит оценки на обоих площадках. Тогда как люди, ставящие оценки фильмам, могут посмотреть его как на нетфликсе, так и в кинотеатрах.

Из этого можно сделать вывод, что аудитория, которая смотрит сериалы очень похожа (в отличии от аудитории, которая оценивает фильмы), и при выпуске нового сериала можно расчитывать на равные оценки как на imdb, так и на netflix.
